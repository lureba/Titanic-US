# Objective
 The goal of this project is to perform an exploratory data analysis (EDA) and build a robust predictive model to predict the survival of Titanic passengers. The analysis combines visual storytelling with technical insights into the factors that influenced survival, such as gender, social class, age, fare, and embarkation port.

## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly as ply
```

```python
df = pd.read_csv("TitanicDF.csv")
```

## Final Insights
Based on our exploratory analysis, we identified clear patterns in the survival of Titanic passengers:

- Gender and Age: Women and children had significantly higher survival rates compared to men. We observed that from the age of 15 onward, the number of deaths exceeded the number of survivors, and this trend persisted until around the ages of 5055, when it reversed again.

- Age Distribution and Social Behavior: There was a higher proportion of young women compared to young men. Combined with the societal norms of the time, prioritizing the rescue of women and children likely contributed to the higher female survival rate.

- Social Class: Social class had a considerable impact on survival chances. First-class passengers, in particular, had a much higher probability of survival, which also correlated with ticket prices  the higher the fare, the greater the likelihood of survival.

- Embarkation Port: The embarkation point was also relevant. Passengers who boarded in Cherbourg had higher survival rates, likely due to the greater presence of high-class passengers embarking from this port.
